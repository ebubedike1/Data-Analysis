```python
# Task 4

import pandas as pd

netflix_df = pd.read_csv(r"C:\Users\820 G3\Desktop\Datasets\netflix_data.csv")
netflix_df.head(5)

```

```python
# Task 5 Subsetting rows and columns
netflix_df_movies_only = netflix_df[netflix_df["type"] == "Movie"]

netflix_df_movies_only.head(5)
```

```python
# Subsetting columns

netflix_movies_col_subset = netflix_df_movies_only[["title", "country", "genre", "release_year", "duration"]]
netflix_movies_col_subset.head(5)
```

```python
# Task 6 -Scatter plot showing release year and duration

plt.scatter(netflix_movies_col_subset["release_year"], netflix_movies_col_subset["duration"])
plt.xlabel("release_year")
plt.ylabel("duration")
plt.title("Year of release")
plt.show()
```

```python
# Task 7 Movies with duration fewer than 60 minutes

short_movies = netflix_movies_col_subset[netflix_movies_col_subset["duration"] < 60]

short_movies.head(20)
```
